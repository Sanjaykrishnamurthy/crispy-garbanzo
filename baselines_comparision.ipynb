```python
import pandas as pd
```

```python
# The data should be in this format. Then we calculate the percentage improvement of one baseline over the other in the next steps.
df = pd.read_clipboard()
df.head()
```

```python
df.set_index('Model', inplace=True)
```

```python
def improvement_percent(df, model_a, model_b):
    df = df.T
    improvement_percent = (df[model_b] - df[model_a]) / df[model_a] * 100
    return improvement_percent.values
```

```python
models_all = df.index
metrics = df.columns
df_compare = []
for model_a in models_all:
    df_model = pd.DataFrame(metrics, columns=['Metric'])
    df_model['Model'] = model_a
    for k, model_b in enumerate(models_all):
        # print(model_a, model_b)
        df_model[model_b] = improvement_percent(df, model_a, model_b)
    print(df_model)
    df_compare.append(df_model)
    # print() 
```

```python
df_compare = pd.concat(df_compare)
df_compare.to_clipboard(index=False)
```
